```python
def cost(x, y, m):
    M = x.shape[0]
    #print('M ', M)
    cost = 0
    for i in range(M):
        cost += ((y[i] - ((m * x[i, :]).sum()))**2)/M
        #print("in ",cost)
    return cost
```

```python
def compute_slope(x, y, m, j):
    M = x.shape[0]
    m_slope = 0
    for i in range(M):
        m_slope += ((-2/M)*(y[i] - (m * x[i, :]).sum())*x[i, j])
    return m_slope
```

```python
def gradient_descent_step(x, y, learning_rate, m):
    M = x.shape[0]
    N = x.shape[1]
    m_slope = np.array([0]*N)
    
    for j in range(N):
        m_slope[j] = compute_slope(x, y, m, j)
        
    m = m - learning_rate * m_slope
    #print(m)
    return m
```

```python
def gradient_descent(x, y, learning_rate, num_iterations):
    M = x.shape[0]
    N = x.shape[1]
    m = np.array([0]*N)
    
    for i in range(num_iterations):
        m = gradient_descent_step(x, y, learning_rate, m)
        print("Cost ", cost(x, y, m))
    return m
```
